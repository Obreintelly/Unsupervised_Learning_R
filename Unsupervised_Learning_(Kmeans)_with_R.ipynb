**Description of Columns**

The dataset consists of 10 numerical and 8 categorical attributes. 

The 'Revenue' attribute can be used as the class label.

"Administrative", "Administrative Duration", "Informational", "Informational Duration", "Product Related" and "Product Related Duration" represents the number of different types of pages visited by the visitor in that session and total time spent in each of these page categories. 

The values of these features are derived from the URL information of the pages visited by the user and updated in real-time when a user takes an action, e.g. moving from one page to another. 

The "Bounce Rate", "Exit Rate" and "Page Value" features represent the metrics measured by "Google Analytics" for each page in the e-commerce site. 

The value of the "Bounce Rate" feature for a web page refers to the percentage of visitors who enter the site from that page and then leave ("bounce") without triggering any other requests to the analytics server during that session. 

The value of the "Exit Rate" feature for a specific web page is calculated as for all pageviews to the page, the percentage that was the last in the session.

The "Page Value" feature represents the average value for a web page that a user visited before completing an e-commerce transaction. 

The "Special Day" feature indicates the closeness of the site visiting time to a specific special day (e.g. Mother’s Day, Valentine's Day) in which the sessions are more likely to be finalized with the transaction. 

The value of this attribute is determined by considering the dynamics of e-commerce such as the duration between the order date and delivery date. For example, for Valentina’s day, this value takes a nonzero value between February 2 and February 12, zero before and after this date unless it is close to another special day, and its maximum value of 1 on February 8. 


The dataset also includes the operating system, browser, region, traffic type, visitor type as returning or new visitor, a Boolean value indicating whether the date of the visit is weekend, and month of the year.

# **K-Means Clustering**

# **Relevant Libraries**

```r
#Packages and Libraries 
install.packages("caret")
library(caret)
```

```r
#Kmeans is an unsupervised machine learning. We there won't need the label
#column as kmeans will help us determine what the labels would be

#Separating the dependent variable from the independent 
kira_indep <- kira[, 1:17]
```

```r
head(kira_indep)
```

```r
unique(kira_indep$Weekend)
```

```r
#Categorical columns Encoding 

#Month type

kira_indep$Month = ifelse(kira_indep$Month == "Aug", 0,
            ifelse(kira_indep$Month == "Dec", 1,
            ifelse(kira_indep$Month == "Feb", 2,
            ifelse(kira_indep$Month == "Jul", 3,
            ifelse(kira_indep$Month == "June", 4,
            ifelse(kira_indep$Month == "Mar", 5,
            ifelse(kira_indep$Month == "May", 6,
            ifelse(kira_indep$Month == "Nov", 7,
            ifelse(kira_indep$Month == "Oct", 8, 9)))))))))

#Visitor Type Column

kira_indep$VisitorType = ifelse(kira_indep$VisitorType == "New_Visitor", 0,
                  ifelse(kira_indep$VisitorType == "Returning_Visitor", 1,2))

#Weekend column
kira_indep$Weekend = ifelse(kira_indep$Weekend == "False", 0, 1)
```

```r
#Confirming Structure of the data after encoding 
str(kira_indep)
```

# **Normalizing the data**

```r
prep <- preProcess(kira_indep, method=c("range"))
norm <- predict(prep, kira_indep)
```

# **Determining No of Clusters**

# **Relevant Library and Package**

```r
install.packages(c("factoextra",  "NbClust", "mlbench"))

library(mlbench)
library(factoextra)
library(NbClust)
```

```r
#Determining number of clusters for the kmeans model
set.seed(42)
fviz_nbclust(norm, FUN =kmeans, method = "wss") +
theme(axis.text = element_text(size=18),
          axis.title = element_text(size = 18),
          plot.title = element_text(hjust = 0.5, size = 20))
```

```r
#Dropping the weekend column to see impact on the clusterplot
norm <- select(norm, -Weekend)
```

```r
head(norm)
```

```r
#Determining number of clusters for the kmeans model after dropping Weekend
set.seed(42)
fviz_nbclust(norm, FUN =kmeans, method = "wss") +
theme(axis.text = element_text(size=18),
          axis.title = element_text(size = 18),
          plot.title = element_text(hjust = 0.5, size = 20))
```

```r
#Checking with the Silhoette Method
fviz_nbclust(norm, FUNcluster = kmeans, method = "silhouette") +
theme(axis.text = element_text(size=18),
          axis.title = element_text(size = 18),
          plot.title = element_text(hjust = 0.5, size = 20))
```

```r
#The optimal number of k with the two approaches is approx 2
```

```r
#Applying the k-means clustering algorithm

k_means <-kmeans(norm, 2, nstart=20)
```

```r
#Total Points per cluster
k_means$size
```

```r
#Cluster centre datapoints by columns
k_means$centers
```

```r

# visualize presentation of the clusters
fviz_cluster(k_means, data = norm) +
theme(axis.text = element_text(size=18),
              axis.title = element_text(size = 18),
              plot.title = element_text(hjust = 0.5, size = 22),
              legend.title = element_text(size=18),
              legend.text = element_text(size=18))
```

# **Challenging the Solution**

```r
#For hierarchical clustering, we will use the dist() function to calculate
#what the Euclidean distance between variables will be

d <- dist(norm, method= "euclidean")

#We will then cluster using hierarchy by leverage Ward methodology
res.hc <- hclust(d, method = "ward.D2" )
```

```r
#Plotting the resulting Dendogram
plot(res.hc, cex = 1, hang= -1, cex.main= 1.75, cex.lab=1.5, cex.axis=1.2)

```

```r
# Creating a border around the clusters
plot(res.hc, cex = 1, hang= -1, cex.main= 1.75, cex.lab=1.5, cex.axis=1.2)
rect.hclust(res.hc, k = 2, border = 2:5)
abline(h = 3, col = 'gold')
```

```r
#Number of records per cluster
cut <- cutree(res.hc, k=2)
table(cut)
```

```r
#How does the model perform
table(cut, kira$Revenue)
```

```r
fviz_cluster(list(data = norm, cluster = cut)) +
theme(axis.text = element_text(size=18),
              axis.title = element_text(size = 18),
              plot.title = element_text(hjust = 0.5, size = 22),
              legend.title = element_text(size=18),
              legend.text = element_text(size=18))
```
